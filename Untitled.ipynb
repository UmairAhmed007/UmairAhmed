```python
def arithmetic_operations():
    
    col1 = ds["Age"].head(5).to_numpy()  
    col2 = ds["MonthlyIncome"].head(5).to_numpy()  

    
    arr = np.array(col1)  
    arr2 = np.array(col2)  

    print("Array 1 (Age):", arr)
    print("Array 2 (MonthlyIncome):", arr2)

    addition = arr + arr2
    subtraction = arr - arr2
    multiplication = arr * arr2
    division = arr / arr2  

    print("Addition:", addition)
    print("Subtraction:", subtraction)
    print("Multiplication:", multiplication)
    print("Division:", division)

arithmetic_operations()
```

```python
def indexing_slicing_examples():
    arr = ds["Age"].head(10).to_numpy()  
    
    slice1 = arr[5:8]  
    arr[5:8] = 12  

    arr2d = ds[["Age", "MonthlyIncome", "DistanceFromHome"]].head(3).to_numpy()  
    last_column = arr2d[:, -1] 
    middle_array_last_two = arr2d[1, -2:]  

    return slice1, arr, arr2d, last_column, middle_array_last_two

slice1, arr, arr2d, last_column, middle_array_last_two = indexing_slicing_examples()

print("Slice 1:", slice1)
print("Array after broadcasting:", arr)
print("2D Array:\n", arr2d)
print("Last Column:", last_column)
print("Middle Array Last Two Elements:", middle_array_last_two)
```

```python

def analyze_dataset(data):
    age_array = data["Age"].to_numpy()  
    income_array = data["MonthlyIncome"].to_numpy()  
    distance_array = data["DistanceFromHome"].to_numpy()  

    mean_age = np.mean(age_array)
    max_income = np.max(income_array)
    mean_distance = np.mean(distance_array)
    min_distance = np.min(distance_array)
    distance_below_mean = distance_array < mean_distance  

    
    return mean_age, max_income, mean_distance, min_distance, distance_below_mean

mean_age, max_income, mean_distance, min_distance, distance_below_mean = analyze_dataset(ds)

print("Mean Age:", mean_age)
print("Max Monthly Income:", max_income)
print("Mean Distance from Home:", mean_distance)
print("Min Distance from Home:", min_distance)
print("Distance Below Mean (Boolean Array):", distance_below_mean)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python



data = pd.read_csv("dataset.csv") 

def line_graph(data, x_col, y_col, num_rows=10):
    x = data[x_col].head(num_rows)
    y = data[y_col].head(num_rows)
    

    plt.plot(x, y, label=f'{x_col} vs {y_col}', marker='o')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.title(f'Line Graph: {x_col} vs {y_col}')
    plt.legend()
    plt.grid(True)
    plt.show()

line_graph(data, "Age", "DistanceFromHome", num_rows=15)  #
```

```python
def bar_chart(data, column_name):
    value_counts = data[column_name].value_counts()
    labels = value_counts.index  
    values = value_counts.values  

    plt.bar(labels, values, color='b')
    plt.xlabel(column_name)
    plt.ylabel('Count')
    plt.title(f'Bar Chart for {column_name}')
    plt.xticks(rotation=45, ha='right')  
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

bar_chart(data, "Department")
```

```python
def scatter_plot(data, x_col, y_col, num_rows=None):
 
    x = data[x_col].head(num_rows) if num_rows else data[x_col]
    y = data[y_col].head(num_rows) if num_rows else data[y_col]

    plt.scatter(x, y, color='blue', label=f'{x_col} vs {y_col}', s=20, alpha=0.7)
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.title(f'Scatter Plot: {x_col} vs {y_col}')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

scatter_plot(data, "YearsAtCompany", "MonthlyIncome", num_rows=50)
```

```python
def histogram(data, column_name, bins=10):

    column_data = data[column_name].dropna() 
    
    plt.hist(column_data, bins=bins, color='orange', alpha=0.7, rwidth=0.85)
    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.title(f'Histogram for {column_name}')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

histogram(data, "DistanceFromHome", bins=15)
```

```python
def pie_chart(data, column_name):
   
    value_counts = data[column_name].value_counts()
    labels = value_counts.index  
    sizes = value_counts.values  
    colors = plt.cm.Paired.colors[:len(labels)]  
    
    plt.pie(sizes, labels=labels, colors=colors, startangle=90, autopct='%1.1f%%', shadow=True)
    plt.title(f'Pie Chart for {column_name}')
    plt.show()


pie_chart(data, "BusinessTravel")
```

```python
def customized_plot(data, x_col, y_col, num_rows=10):
    x = data[x_col].head(num_rows)
    y = data[y_col].head(num_rows)
    
    plt.plot(
        x, y, color='green', linestyle='dashed', linewidth=2,
        marker='o', markerfacecolor='blue', markersize=8
    )
    plt.ylim(0, y.max() + (0.1 * y.max()))  
    plt.xlim(x.min() - 1, x.max() + 1)
    plt.xlabel(x_col)  
    plt.ylabel(y_col)
    plt.title(f'Customized Line Graph: {x_col} vs {y_col}')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()


customized_plot(data, "YearsAtCompany", "MonthlyIncome", num_rows=15) 
```

```python
def curve_plot(data, column_name, num_rows=100):
  
    y = data[column_name].head(num_rows).to_numpy() 
    x = np.arange(len(y))  
  
    plt.plot(x, y, label=f'{column_name} Curve', color='purple')
    plt.xlabel('Index')
    plt.ylabel(column_name)
    plt.title(f'Curve Plot for {column_name}')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

curve_plot(data, "MonthlyIncome", num_rows=50)  
```
